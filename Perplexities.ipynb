```python
import subprocess
import re

def get_perplexities(corpus, documents):
    perplexities = subprocess.Popen(["./perplexity.sh", corpus, documents, "DS", "0.75"], cwd="./ngram", stdout=subprocess.PIPE)
    sorted_perplexities = subprocess.check_output(["sort", "-k", "2"], stdin=perplexities.stdout)
    
    return sorted_perplexities

# From the output of get_perplexity(corpus, documents), 
# create a nice python dictionary of the form {document_name : perplexity}
def output2dict(output, delim=":\s+"):
    lookup = {}
    
    lines = output.splitlines()
    for line in lines:
        item = re.split(delim,line)

        key = item[0]
        ppl = item[1]
        
        lookup[key] = ppl
    
    return lookup
```

```python
episode_perplexities = get_perplexities("data/allep", "data/episodes")

print(episode_perplexities)
episode_dict = output2dict(episode_perplexities)
```

```python
season_perplexities = ""
for i in range(1,10):
    season_perplexity = get_perplexities("data/seasons/nots0{}".format(i), "data/seasons/szn0{}".format(i))
    season_perplexities += season_perplexity

print(season_perplexities)
season_dict = output2dict(season_perplexities)
```

```python
author_perplexities = ""
authors = ["paul", "greg", "mindy", "jennifer"]
for author in authors:
    author_perplexity = get_perplexities("data/writers/not{}".format(author), "data/writers/{}".format(author))
    author_perplexities += author_perplexity
    
print(author_perplexities)
author_dict = output2dict(author_perplexities)
```

```python
output = ""
with open('ratings', 'r') as f:
    output = f.read()

episode_ratings = output2dict(output, "\s+")
```

```python
import csv
def write_ratings(ratings_dict, perplexity_dict, filename):
    with open(filename, 'wb') as csvfile:
        writer = csv.writer(csvfile, delimiter=',')
        writer.writerow(["episode", "rating", "perplexity"])
        
        for ep in perplexity_dict.keys():
            writer.writerow([ep, ratings_dict[ep], perplexity_dict[ep]])
            
write_ratings(episode_ratings, episode_dict, "episode_ratings.csv")
```
